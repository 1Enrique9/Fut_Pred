```python
import pandas as pd
import os
from pathlib import Path

# === Ruta base donde est√°n las carpetas de los equipos ===
base_path = Path(r"D:\Fut_Pred\Datos_jugadores\2024-2025\Inglaterra")  # <-- c√°mbiala a la tuya

# === Carpeta donde se guardar√°n los archivos combinados ===
output_folder = base_path / "Combinados"
output_folder.mkdir(exist_ok=True)

# === Recorre todas las carpetas dentro de la ruta base ===
for folder in base_path.iterdir():
    if folder.is_dir():
        excel_files = list(folder.glob("*.xlsx"))

        if not excel_files:
            print(f"‚ö†Ô∏è No hay archivos Excel en {folder.name}, se omite.")
            continue

        print(f"üìÇ Combinando archivos en: {folder.name} ({len(excel_files)} archivos)")

        # Leer y concatenar
        dfs = []
        for f in excel_files:
            df = pd.read_excel(f, sheet_name=0, engine="openpyxl")
            dfs.append(df)

        combined_df = pd.concat(dfs, ignore_index=True)

        # Guardar con el nombre de la carpeta
        output_path = output_folder / f"{folder.name}.xlsx"
        combined_df.to_excel(output_path, index=False)

        print(f"‚úÖ Guardado: {output_path}")

print("\nüéâ ¬°Proceso terminado!")
```
