```python
import pandas as pd
import os
from pathlib import Path

# === Ruta base donde estÃ¡n las carpetas de los equipos ===
base_path = Path(r"D:\Fut_Pred\Datos_jugadores\2024-2025\Inglaterra")  # <-- cÃ¡mbiala a la tuya

# === Carpeta donde se guardarÃ¡n los archivos combinados ===
output_folder = base_path / "Combinados"
output_folder.mkdir(exist_ok=True)

# === Recorre todas las carpetas dentro de la ruta base ===
for folder in base_path.iterdir():
    if folder.is_dir():
        csv_files = list(folder.glob("*.csv"))

        if not csv_files:
            print(f"âš ï¸ No hay archivos CSV en {folder.name}, se omite.")
            continue

        print(f"ðŸ“‚ Combinando archivos en: {folder.name} ({len(csv_files)} archivos)")

        # Leer y concatenar todos los CSV
        dfs = []
        for f in csv_files:
            df = pd.read_csv(f)
            dfs.append(df)

        combined_df = pd.concat(dfs, ignore_index=True)

        # Guardar con el nombre de la carpeta
        output_path = output_folder / f"{folder.name}.csv"
        combined_df.to_csv(output_path, index=False, encoding="utf-8-sig")

        print(f"âœ… Guardado: {output_path}")

print("\nðŸŽ‰ Â¡Proceso terminado!")
```
