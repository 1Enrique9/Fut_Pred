### Equipos 

Evitar los duplicados, o dobles tablas 

```python
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS bronze_teams;"))
    conn.commit()
print("Tabla 'bronze_teams' eliminada correctamente.\n")
```

Espa√±a 

```python
#  Carpeta con los CSV 
folder_path = r"D:\Fut_Pred\Data_equipos\2024-2025\Espa√±a"
csv_files = glob(os.path.join(folder_path, "*.csv"))
print(f" Se encontraron {len(csv_files)} archivos en {folder_path}\n")

#  Funci√≥n para limpiar nombres 
def clean_column_name(col):
    col = str(col).strip()
    col = re.sub(r"[^\w]+", "_", col)  
    col = re.sub(r"__+", "_", col)
    col = col.strip("_")
    return col[:60]

# Cargar archivos 
for csv_path in csv_files:
    try:
        df = pd.read_csv(csv_path, sep=",", encoding="utf-8")
        df.columns = [clean_column_name(c) for c in df.columns]

        # Eliminar duplicados de encabezados
        df = df.loc[:, ~df.columns.duplicated()]

        # Metadatos
        df["league"] = "Laliga"
        df["season"] = "2024-2025"
        df["team_name"] = os.path.splitext(os.path.basename(csv_path))[0]
        df["file_source"] = os.path.basename(csv_path)

        # Subir a MySQL
        df.to_sql("bronze_teams", engine, if_exists="append", index=False, method="multi", chunksize=1000)
        print(f" Cargado: {os.path.basename(csv_path)} ({len(df)} filas)")

    except Exception as e:
        print(f" Error al cargar {os.path.basename(csv_path)} ‚Üí {e}")

print("\nüèÅ Proceso completado: equipos de Espa√±a cargados correctamente en 'bronze_teams'")
```
